### 1.2 Region growing
Add your implementation and discussion

````python
# define a funcion that return 8 neighbors of a start point that satisfied the set threshold
def find_neighbor(image, threshold, start_x, start_y):
    img_len = image.shape[0]
    img_width = image.shape[1]
    #new_image=np.zeros((img_len,img_width))
    new_x=np.arange(start_x-1,start_x + 2)
    new_y=np.arange(start_y-1,start_y + 2)
    candidate=[]
    for temp_x in new_x:
        for temp_y in new_y:
            # discard neighbor points that are beyond the size of image
            if temp_x<0 or temp_x>=img_len:
                continue
            if temp_y<0 or temp_y>=img_width:
                continue
            # only get new neighbor points that don't exist in our candidate list
            if [temp_x, temp_y] in candidate:
                continue
            # no need to consider the center point itself
            if temp_x==start_x and temp_y==start_y:
                continue
            if image[temp_x][temp_y] > threshold:
                candidate.append([temp_x,temp_y])
    return candidate
````

````python
def region_growing(image,threshold,start_x,start_y):
    # initialization
    img_len = image.shape[0]
    img_width = image.shape[1]
    new_image=np.zeros((img_len,img_width))
    candidates=[[start_x,start_y]]
    
    while len(candidates)>0:
        temp_point = candidates.pop()
        candidate_neigbor=find_neighbor(image, threshold, temp_point[0], temp_point[1])
        for temp_neibor in candidate_neigbor:
            if new_image[temp_neibor[0]][temp_neibor[1]]==0:
                # if the candidate neignbor is not in current list, we append the point
                # and set the corresponding pixel in image to 1
                new_image[temp_neibor[0]][temp_neibor[1]]=1
                candidates.append(temp_neibor)
    return new_image
        
````

### Fitting by Gaussian Mixture Model

We use the same thresholding method, but this time the thresholds are not set through obervation and experiments.

Instead, the thresholds are set through the fixed Gaussian models.

Since the orgin histogram shows the pixel distribution has 2 minimums and 2 maximums, we can use 2 Gaussian models to fit the distribution.

````python
all_pixels=np.concatenate((brain_im.ravel().reshape(-1,1)))
# fit the distribution with 2 Gaussians
gau_model = GaussianMixture(n_components = 2, tol=0.0001,max_iter=500)
gau_model.fit(np.expand_dims(all_pixels, 1))
base = np.linspace(np.min(all_pixels), np.max(all_pixels), 500) 
# get the parameters of each Gaussian
for i in range(2):
        temp_mean=gau_model.means_.flatten()[i]
        temp_std=np.sqrt(gau_model.covariances_.flatten())[i]
        temp_weight=gau_model.weights_[i]
        temp_model=scipy.stats.norm(temp_mean, temp_std).pdf(base) * temp_weight
        print('mean {:} = {:}, std {:} = {:}, weight {:} = {:}'.format(str(i+1),str(temp_mean)
                                                                                ,str(i+1), str(temp_std),str(i+1), str(temp_weight)))
````

````python
mix_model = np.exp([gau_model.score_samples(e.reshape(-1, 1)) for e in base]) 
plt.plot(base, mix_model)
````

#### Vocabulary


* **IoU**: Let $A$ be the ground truth area and $B$ the predicted one. the intersection over union (IoU) is given as:
$$IoU(A,B) = \frac{A \cap B}{A \cup B} = \frac{\text{Area of Overlap}}{\text{Area of Union}}$$
Let $T$ be a scalar in the interval $[0, 1]$, then $B$ will be considered as a true positive (TP) if and only if $IoU(A,B) \ge T$. As a results, we can evaluate the performance of our detection with different values of $T$. The larger $T$ is, the more constraining is our condition.


* **Precision**: The precision is given as:
$$precision = \frac{tp}{tp + fp}$$
where $tp$ is the number of true positives and $fp$ the number of false positives. The precision is intuitively the ability of the classifier not to label as positive a sample that is negative.


* **Recall**: The recall is given as:
$$precision = \frac{tp}{tp + fn}$$
where $tp$ is the number of true positives and $fn$ the number of false negatives. The recall is intuitively the ability of the classifier to find all the positive samples.


* **F1-score**: The F1 score can be interpreted as a weighted average of the precision and recall, where an F1 score reaches its best value at 1 and worst score at 0. The relative contribution of precision and recall to the F1 score are equal. The formula for the F1 score is:
$$F1 = \frac{2 \cdot precision \cdot recall}{precision + recall}$$


**FAQ**

Q: What to do if I have multiple ground truth $A_i$ and candidates $B_j$ and want to find the best set $(i,j)$ such that $(i, j) = \arg\max_{i,j} IoU(A_i, B_j)$ ? 

A: Here, we recommend to use a brute force approach. It corresponds to compute $IoU(A, B)$ for every possible pairs (i,j) and keep the pair with the highest $IoU$. Note that a candidate $B_j$ can **only** be matched to **one** ground truth area $A_i$.

```xml
<annotation>
```

The block `object` defines the attributes of the varroa. `name` is the definition of the object class (always varroa). `bndbox` is the definition of the bounding box. To simplify your task, we implemented a function named `parse_file` that returns an array with the bounding box of the varroa expressed as `(x, y, width, heigth)` which corresponds to `(xmin, ymin, xmax-xmin, ymax-ymin)`

````python
import xml.etree.ElementTree as ET

def parse_file(filename):
    """ Parse a PASCAL VOC xml file """
    tree = ET.parse(filename)
    objects = []
    for obj in tree.findall('object'):
        obj_struct = {}
        obj_struct['name'] = obj.find('name').text
        bbox = obj.find('bndbox')
        obj_struct['bbox'] = [int(float(bbox.find('xmin').text)),
                              int(float(bbox.find('ymin').text)),
                              int(float(bbox.find('xmax').text))-int(float(bbox.find('xmin').text)),
                              int(float(bbox.find('ymax').text))-int(float(bbox.find('ymin').text))]
        objects.append(obj_struct)

    return objects
````

### 2.2 Count the number of Varroa on each picture

Add your implementation and discussion

Hint: `skimage.measure.label` might be useful to label the thresholded image

According to 4 varroa images, we find the varroa objects are in dark color with flat shape and large size. Among the images, there are also other objects (usually yello) and shallow-color background. Thus, our method are as follows:

1. get the 2 potential thresholds and extract pixel larger than smaller threshold

2. get the labels of bounding boxes 

3. filter the bounding boxes that satisfy the varroa feature (large area and not large eccentricity)

4. visualize the predicted bounding box in image, and compare them with ground truth

````python
# define a function that predict the bounding box 
# and visualize the performance of prediction vs ground truth 
def count_varroa(image,min_area,max_eccentricity,temp_annotations,ax_1,ax_2):
    # visualze the ground truth bounding box like in 2.1 
    # and display the number of bounding boxes in ground truth
    ax_1.imshow(image)
    for anno in temp_annotations:
        rect = patches.Rectangle((anno['bbox'][0], anno['bbox'][1]), anno['bbox'][2], anno['bbox'][3],
                linewidth=1,edgecolor='r',facecolor='none')
        
        ax_1.add_patch(rect)
    ground_truth_count=len(temp_annotations)
    ax_1.set_title('Ground Truth: '+str(ground_truth_count))
    
    gray_image=skimage.color.rgb2gray(image)*256
    # automatically find the minimal threshold
    temp_threshold=skimage.filters.threshold_minimum(gray_image)
    # get the labels that's larger than the set threshold, and store in candidates
    varroa_labels = skimage.measure.label(gray_image>temp_threshold,background=True)
    candidates=skimage.measure.regionprops(varroa_labels)
    varroa_candidate=[]
    # get the final candidate(varroa_candidate) box that satisfied the area and eccentricity
    for candidate in candidates:
        #print(str(candidate.area))
        if candidate.area>min_area and candidate.eccentricity<max_eccentricity:
            varroa_candidate.append(candidate)
            
    ax_2.imshow(image)
    # visualize bounding box of varroa_candidate along with original image
    for candidate in varroa_candidate:
        temp_bbox = [candidate['bbox'][0], candidate['bbox'][1], candidate['bbox'][2] - candidate['bbox'][0], 
                candidate['bbox'][3] - candidate['bbox'][1]]
        rect = patches.Rectangle((temp_bbox[1], temp_bbox[0]), temp_bbox[2], temp_bbox[3],
                                  linewidth=1,edgecolor='r',facecolor='none')
        ax_2.add_patch(rect)
    # print the number of predicted bounding box in the plot title
    ax_2.set_title('Our Prediction: '+str(len(varroa_candidate)))
    return varroa_candidate
````

````python
# compute the intersection of 2 bounding boxes according to the theoretical definition
def compute_intersection(box_gt,box_p):
    intersection=0
    temp_x = min(box_gt[0] + box_gt[2], box_p[0] + box_p[2]) - max(box_gt[0], box_p[0])
    temp_y = min(box_gt[1] + box_gt[3], box_p[1] + box_p[3]) - max(box_gt[1], box_p[1])
    if temp_x>=0 and temp_y>=0:
        intersection=temp_x*temp_y
    return intersection
````

````python
# compute the IOU of 2 bounding boxes according to the theoretical definition
def compute_IOU(box_gt,box_p):
    temp_iou=0
    intersection=compute_intersection(box_gt,box_p)
    # compute the union of 2 bounding boxes according to the theoretical definition
    union = box_gt[2] * box_gt[3] + box_p[2] * box_p[3] - intersection
    if intersection!=0:
        temp_iou=intersection/union
    return temp_iou
````

````python
# define a function that 
def display_IOU(annotations,predictions):
    ious=[]
    # extract the bounding box from annotations
    ann_boxes=[temp_ann['bbox'] for temp_ann in annotations]
    # extract the bounding box from predicted regionprop
    pred_boxes=[]
    for temp_pred in predictions:
        temp_pred_box = temp_pred.bbox
        temp_box = [temp_pred_box[1],temp_pred_box[0], temp_pred_box[3] - temp_pred_box[1],temp_pred_box[2] - temp_pred_box[0]]
        pred_boxes.append(temp_box)
        
    # for each predicted bounding box, find an annotated box that matches it best
    # i.e. most overlapping area
    for pred in pred_boxes:
        temp_match=[]
        for ann in ann_boxes:
            temp_intersection=compute_intersection(ann,pred)
            temp_match.append(temp_intersection)
        if len(temp_match)>0:
            # best matched annotated box has max length
            best_match=max(temp_match)
            temp_index=temp_match.index(best_match)  
            match_annoation=ann_boxes[temp_index]
            # find the best match box and calculate the IoU for each predicted bounding box
            temp_iou=compute_IOU(match_annoation,pred)
            ious.append(temp_iou)
    # visualize the IoU value list by printing and histogram
    print('IOU Values of Our Predictions vs Ground Truth')
    print(str(ious))
    plt.hist(ious,bins=30)
    plt.show()
    return ious
    
````

### 2.4 Precision, Recall and F1-score at two different IoU thresholds

Add your implementation and discussion

````python
# define a function that calculate the visualze the metrics
def calculate_statistics(annotations,predictions,ious,threshold):
  
    # get TP.FN,FP from definitation
    tp=np.sum([temp_iou>threshold for temp_iou in ious])
    fn=len(annotations)-tp
    fp=len(predictions)-tp
    
    # calucalte precision, recall and f1 score by mathematical definition
    precision=recall=f1=0
    if tp + fp !=0:
        precision=tp/(tp + fp)
    if tp + fn !=0:
        recall=tp/(tp + fn)
    if tp + fp !=0 or tp + fn !=0:
        f1=2*precision*recall/(precision + recall)
    # visualize the metric
    print('when threshold = '+str(threshold))
    print ('precision = '+str(precision)+' , recall = '+str(recall)+' , f1 score = '+str(f1))
    return precision,recall,f1
````
