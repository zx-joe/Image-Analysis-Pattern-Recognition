### 1.2 Fourier descriptors
Add your implementation and discussion

Pipeline:
1. for each image, find its largest contour
2. calucate fourier transform based on largest contour
3. get f1,f2,f3 according to fourier transform
4. plot the scatter plot based on calculated f1,f2 and f3

```python
#this function return the max contour of image
# and the corresponding matrix
from skimage import measure
def find_max_contour(image,threshold=100):
    # get all contours through 'measure.find_contours'
    contours = measure.find_contours(image, threshold)
    # get the length of all contours 
    # and find the contour with largest length
    contours_len=[temp_contour.shape[0] for temp_contour in contours]
    max_index=np.argmax(contours_len)
    max_coutour=contours[max_index]
    # initialize a zero matrix
    # and fill the matrix by max contour
    contour_show=np.zeros(image.shape)
    for x, y in zip(max_coutour[:,1], max_coutour[:,0]):
        contour_show[int(round(y)),int(round(x))] = 1
    return max_coutour,contour_show
        
```

```python
temp_image=zeros_im[0]
temp_max_contour,temp_contour_show=find_max_contour(temp_image)
fig, ax = plt.subplots(1,2, figsize=(5,5))
ax[0].imshow(temp_image, cmap='gray');
ax[0].set_title('Original image')
ax[1].imshow(temp_contour_show, cmap='gray')
ax[1].plot(temp_max_contour[:, 1], temp_max_contour[:, 0])
ax[1].set_title('Contour ');
```

We can take a look at the performance of max_conout from the first image in zeros_im.

It can be observed that visually the contour is successfully extracted.

Afterwards, we will conduct Fourier Transform based on the max contour.

```python
# define a function that return fourier vectors of image
def find_fourier_descriptor(image):
    # get max contour
    max_contour,_=find_max_contour(image)
    # calculate the fourier transform by fft
    temp_vector = max_contour[:,1] + 1j*max_contour[:,0]
    fft_vector = np.fft.fft(temp_vector)
    return fft_vector
    
```

```python
# this function returns 2 different values after fourier transform
# if amp=1 (default), return the 2 values of features vectors with respect to reference feature
# the default of 2 feature vectors are 1 and 2
# the default of reference feature is 0
# if amp=0, return the 2 absolute values of features vectors 
def fourier_2D_matrix(images,index_1=1,index_2=2,amp=1,index_ref=0):
    abs_list=[]
    for temp_image in images:
        temp_fft_vector=find_fourier_descriptor(temp_image)
        if amp==1:
            temp_abs_1=abs(temp_fft_vector[index_1])/abs(temp_fft_vector[index_ref])
            temp_abs_2=abs(temp_fft_vector[index_2])/abs(temp_fft_vector[index_ref])
        else:
            temp_abs_1=abs(temp_fft_vector[index_1])
            temp_abs_2=abs(temp_fft_vector[index_2])
        abs_list.append([temp_abs_1,temp_abs_2])
    return abs_list
```

```python
# plot the 2D scatter plot of 2 feature vecotrs of 2 image sets
def plot_fourier_2D(images_1,images_2,temp_amp,temp_ax):
    # get the list from the calculated fourier matrix
    matrix_1=np.array(fourier_2D_matrix(images_1,amp=temp_amp))
    matrix_2=np.array(fourier_2D_matrix(images_2,amp=temp_amp))
    # plot 2D scatter plot according to fourier matrix
    temp_ax.scatter(matrix_1[:,0],matrix_1[:,1], c = 'red',label='zeros')
    temp_ax.scatter(matrix_2[:,0], matrix_2[:,1], c = 'green',label='ones')
    # according to different input temp_amp, change corresponding labels
    if temp_amp==1:
        temp_ax.set_xlabel('abs(f1/f0)')
        temp_ax.set_ylabel('abs(f2/f0)')
        temp_ax.legend()
    else:
        temp_ax.set_xlabel('abs(f1)')
        temp_ax.set_ylabel('abs(f2)')
        temp_ax.legend()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
plot_fourier_2D(zeros_im,ones_im,1,ax)
```

It can be observed that the vectors of the "0"'s cluster in right part of the plane and those of the "1"'s in on the left.

For most images of 0 or 1, inter-class variance is larger than intra-class variance.

Now we take a look at 3 vectors calculated from Fourier transform and viualize the corresponding 3D plot.

```python
# this function returns 3 different values after fourier transform
# if amp=1 (default), return the 3 values of features vectors with respect to reference feature
# the default of 2 feature vectors are 1,2 and 3
# the default of reference feature is 0
# if amp=0, return the 3 absolute values of features vectors 
def fourier_3D_matrix(images,index_1=1,index_2=2,index_3=3,amp=1,index_ref=0):
    abs_list=[]
    for temp_image in images:
        temp_fft_vector=find_fourier_descriptor(temp_image)
        if amp==1:
            temp_abs_1=abs(temp_fft_vector[index_1])/abs(temp_fft_vector[index_ref])
            temp_abs_2=abs(temp_fft_vector[index_2])/abs(temp_fft_vector[index_ref])
            temp_abs_3=abs(temp_fft_vector[index_3])/abs(temp_fft_vector[index_ref])
        else:
            temp_abs_1=abs(temp_fft_vector[index_1])
            temp_abs_2=abs(temp_fft_vector[index_2])
            temp_abs_3=abs(temp_fft_vector[index_3])
        abs_list.append([temp_abs_1,temp_abs_2,temp_abs_3])
    return abs_list
```

```python
# import package for 3D plotting
from mpl_toolkits.mplot3d import Axes3D

# plot the 3D scatter plot of 3 feature vecotrs of image sets
def plot_fourier_3D(images_1,images_2,temp_amp,temp_ax):
    # get the list from the calculated fourier matrix
    matrix_1=np.array(fourier_3D_matrix(images_1,amp=temp_amp))
    matrix_2=np.array(fourier_3D_matrix(images_2,amp=temp_amp))
    # plot 3D scatter plot according to fourier matrix
    temp_ax.scatter(matrix_1[:,0], matrix_1[:,1], matrix_1[:,2], c='red',label='zero')  
    temp_ax.scatter(matrix_2[:,0], matrix_2[:,1], matrix_2[:,2], c='green',label='one')
    # according to different input temp_amp, change corresponding labels
    if temp_amp==1:
        temp_ax.set_xlabel('abs(f1/f0)')
        temp_ax.set_ylabel('abs(f2/f0)')
        temp_ax.set_zlabel('abs(f3/f0)')
        temp_ax.legend()
    else:
        temp_ax.set_xlabel('abs(f1)')
        temp_ax.set_ylabel('abs(f2)')
        temp_ax.set_zlabel('abs(f3)')
        temp_ax.legend()
```

Now, we create some rotations of original images.

```python
# the function inputs 2 images sets 
# and num that refers to the number of produced rotations for each image
# the function output the combination of orginal image sets and rotated image sets
def create_image_rotation(images_1,images_2,num):
    angle_list=[]
    # produce uniform random values from 0 to 360 according to input num
    for i in range(num):
        temp_angle=int(np.random.uniform(0, 360))
        angle_list=angle_list+[temp_angle]
    # for each produced random angle, rotate orginal image with the angle 
    # and add the rotated image sets to original sets
    images_1_rotations = [cv2.warpAffine(image, cv2.getRotationMatrix2D((image.shape[0]/2,image.shape[1]/2), angle, 1.0), image.shape)
                        for image in images_1  for angle in angle_list]
    images_2_rotations = [cv2.warpAffine(image, cv2.getRotationMatrix2D((image.shape[0]/2,image.shape[1]/2), angle, 1.0), image.shape)
                        for image in images_2  for angle in angle_list]
    return images_1_rotations,images_2_rotations
```

```python
# the function inputs 2 images sets 
# the function output the rescaled image sets
def create_image_rescale(images_1,images_2):    
    temp_height, temp_width = images_1.shape[1: 3]
    images_1_rescale = [cv2.resize(temp_img, (2 * temp_height, 2 * temp_width), interpolation = cv2.INTER_CUBIC) for temp_img in images_1]
    images_2_rescale = [cv2.resize(temp_img, (2 * temp_height, 2 * temp_width), interpolation = cv2.INTER_CUBIC) for temp_img in images_2]
    return images_1_rescale,images_2_rescale
```

### 1.3.1 Region-based Descriptors

In this part, we focus on Region-based Descriptors covered in lecture. 

We focus on three intuitive descriptor in Region-based Descriptors.

That is compacity, rectangularity and elongation.

Pipeline:

1. for each input image, get the threshold 
2. get the moment from calculated threhold
5. calculate the elongation value from calucated moment
3. find contours of each image
4. get the area and length of contour
5. calculate the compacity of image
6. calcuate the width and length of bounnding box
7. calculate the rectangularity of image
8. visualize the distribution of 2 image sets based on 3 features.

```python
# the function outputs three features according to input image
# that is , compacity, rectangularity and elongation.
def find_simple_descritpors(image,threshold=100):
    # for each input image, get the threshold 
    _,temp_threshold = cv2.threshold(image,threshold,255,cv2.THRESH_BINARY)
    # get the moment from calculated threhold
    temp_moment = cv2.moments(temp_threshold)
    # calculate the elongation value according to mathematical definition
    temp_num=float(temp_moment['mu20']+temp_moment['mu02']+np.sqrt(4*temp_moment['mu11']**2+(temp_moment['mu20']-temp_moment['mu02'])**2))
    temp_den=float(temp_moment['mu20']+temp_moment['mu02']-np.sqrt(4*temp_moment['mu11']**2+(temp_moment['mu20']-temp_moment['mu02'])**2))
    temp_elongation=temp_num/temp_den
    
    # find contours of each image
    _,temp_contours,_ = cv2.findContours(image,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    # get the area and length of contour
    temp_area = cv2.contourArea(temp_contours[0])
    temp_area=float(temp_area)
    temp_length =cv2.arcLength(temp_contours[0],True)
    # calculate the compacity according to mathematical definition
    temp_compacity=temp_length*temp_length/temp_area
    
    # calcuate the width and length of bounnding box
    _,_,temp_w,temp_h = cv2.boundingRect(temp_contours[0])
    temp_rect_area = temp_w*temp_h
    temp_rect_area=float(temp_rect_area)
    #  calculate the rectangularity according to mathematical definition
    temp_rectangularity = temp_area/temp_rect_area
    
    return temp_compacity,temp_rectangularity,temp_elongation
```

```python
# for image sets, get the list of 3 features 
# that corresponds to each image in it
def find_descritpors_list(images,threshold=100):
    capacity_list=[]
    rectangularity_list=[]
    elongation_list=[]
    for temp_image in images:
        # calculate 3 features of each image and add them to the list
        temp_compacity,temp_rectangularity,temp_elongation=find_simple_descritpors(temp_image,threshold=100)
        capacity_list=capacity_list+[temp_compacity]
        rectangularity_list=rectangularity_list+[temp_rectangularity]
        elongation_list=elongation_list+[temp_elongation]
    return capacity_list,rectangularity_list,elongation_list
```

### 1.3.2 Machine Learning Method (PCA, K-means and DBSCAN)

In this section, we utilizz PCA algorithm to sepeate the 2 sets from orginal multidimensional space to 2 or 3 major feature vecotrs.


After that, in addition to  visualization, we use 2 clustering algorithm in machine learning, K-means and DBSCAN, to verify the performance of feature extraction based on PCA.



Piepline:

1. extract lists of 2 or 3 chosen features based on PCA

2. visualize the feature plottings after PCA

3. use K-means clustering in 3D space to verify the performance of extracted features

4. use DBSCAN clusteringto better verify feature extraction

```python
# import sklearn package for machine learning implementation
from sklearn.decomposition import PCA

# this function return the list of chosen feature vectors after PCA
# input: 2 image sets 
# input: num: number of extracted feature vectors, default 2
# input:f1,f2,f3: chosen index of extracted feature vectors, default: first 3 
# output: corresponding lists of extracted feature vectors
def get_pca_component(images_1,images_2,num=2,f1=1,f2=2,f3=3):
    # flattern each input image sets, for later clustering
    images_1_flat=[temp_image.flatten() for temp_image in images_1]
    images_2_flat=[temp_image.flatten() for temp_image in images_2]
    # merge 2 flatterned sets
    all_flat=images_1_flat+images_2_flat
    # get the decomposition of PCA
    temp_pca=sklearn.decomposition.PCA(n_components = 10)
    # fit the sets
    temp_projections=temp_pca.fit_transform(all_flat)
    # extract the predictions of PCA
    pred_1=temp_projections[: len(images_1_flat)]
    pred_2=temp_projections[len(images_2_flat): ]
    # according to input num, extract lists of 2 or 3 chosen features
    if num==2:
        images_1_f1=pred_1[:,f1-1]
        images_1_f2=pred_1[:,f2-1]
        images_2_f1=pred_2[:,f1-1]
        images_2_f2=pred_2[:,f2-1]
        return images_1_f1,images_1_f2,images_2_f1,images_2_f2
    if num==3:
        images_1_f1=pred_1[:,f1-1]
        images_1_f2=pred_1[:,f2-1]
        images_1_f3=pred_1[:,f3-1]
        images_2_f1=pred_2[:,f1-1]
        images_2_f2=pred_2[:,f2-1]
        images_2_f3=pred_2[:,f3-1]
        return images_1_f1,images_1_f2,images_1_f3,images_2_f1,images_2_f2,images_2_f3
        
```

```python
# helo draw function of extracted features after OCA
def draw_pca(images_1,images_2,temp_ax,num=2,f1=1,f2=2,f3=3,angle=150):
    # plot 2D or 3D according to similar inputs with last function
    if num==2:
        zeros_f1,zeros_f2,ones_f1,ones_f2=get_pca_component(images_1,images_2,f1=f1,f2=f2)
        temp_ax.scatter(zeros_f1,zeros_f2, c='red',label='zeros')  
        temp_ax.scatter(ones_f1,ones_f2, c='green',label='ones')
        temp_ax.set_xlabel('f'+str(f1))
        temp_ax.set_ylabel('f'+str(f2))
        temp_ax.set_title("Original PCA")
        temp_ax.legend()
    if num==3:
        zeros_f1,zeros_f2,zeros_f3,ones_f1,ones_f2,ones_f3=get_pca_component(images_1,images_2,num=3,f1=f1,f2=f2,f3=f3)
        temp_ax.scatter(zeros_f1,zeros_f2, zeros_f3, c='red',label='zeros')  
        temp_ax.scatter(ones_f1,ones_f2,ones_f3, c='green',label='ones')  
        temp_ax.set_xlabel('f'+str(f1))
        temp_ax.set_ylabel('f'+str(f2))
        temp_ax.set_zlabel('f'+str(f3))
        temp_ax.set_title("Original PCA")
        temp_ax.legend()
        for aa in range(0, 360):
            temp_ax.view_init(angle, aa)
            plt.draw()
    
```

Now we draw the 3D plot of first 3 features (default as f1,f2 and f3).

And we implement clustering algorithms after PCA.

```python
# function that produces coordination array
# for later 3D visualization
def get_points(f1,f2,f3):
    points=[]
    for i in range(len(f1)):
        temp_x=f1[i]
        temp_y=f2[i]
        temp_z=f3[i]
        temp_point=[temp_x,temp_y,temp_z]
        points.append(temp_point)
    points = np.array(points)
    return points
```

```python
from sklearn.cluster import KMeans
# based on input image sets
# implement K-means clustering and visualze in 3D plot
def show_kmeans(images_1,images_2,temp_ax,angle=150):
    # get pca components
    zeros_f1,zeros_f2,zeros_f3,ones_f1,ones_f2,ones_f3=get_pca_component(images_1,images_2,num=3)
    # transform in array 
    zeros_points=get_points(zeros_f1,zeros_f2,zeros_f3)
    ones_points=get_points(ones_f1,ones_f2,ones_f3)
    # stack all points for clustering 
    all_points=np.row_stack((zeros_points, ones_points))
    # create kmeans model and fit points
    ml_model = KMeans(n_clusters=2)
    ml_model.fit(all_points)
    # predict upon kmeans and input points
    pred=ml_model.predict(all_points)
    # draw 3D plot according to clustering reulst
    temp_ax.scatter(all_points[:, 0], all_points[:, 1], all_points[:,2], c=pred, s=20, cmap='jet')
    temp_ax.set_xlabel('f1')
    temp_ax.set_ylabel('f2')
    temp_ax.set_zlabel('f3')
    temp_ax.set_title("Kmeans Clustering")
    temp_ax.legend()
    for aa in range(0, 360):
            temp_ax.view_init(angle, aa)
            plt.draw()
```

show original 3D plot after projecting images onto first 3 features(f1,f2,f3) after PCA.

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d') 
draw_pca(zeros_im,ones_im,ax,num=3,angle=60)
```

Now we run K-means on original sets and visualize results.

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d') 
show_kmeans(zeros_im,ones_im,ax,angle=60)
```

The overall clustering performance on original sets are good, except one misclassification point in "1".

The overall clustering performance is not very good. There exist some misclassification in the corners of space.

This may be due to that the clustering algorithm of K-means rely on distance and always converge to globular clusters, which may not be helpful to outliers with large intra-class variance.

Then we try DBSCAN clustering algorithm.

```python
from sklearn.cluster import DBSCAN
# based on input image sets
# implement DBSCAN clustering and visualze in 3D plot
# eps and min_samples arehyperparameters of DBSCAN
# they are default based on emprical tuning
def show_dbscan(images_1,images_2,temp_ax,eps=85,min_samples=20,angle=150):
    # get pca components
    zeros_f1,zeros_f2,zeros_f3,ones_f1,ones_f2,ones_f3=get_pca_component(images_1,images_2,num=3)
    # transform in array 
    zeros_points=get_points(zeros_f1,zeros_f2,zeros_f3)
    ones_points=get_points(ones_f1,ones_f2,ones_f3)
    # stack all points for clustering 
    all_points=np.row_stack((zeros_points, ones_points))
    # create DBSCAB model and fit points
    ml_model = DBSCAN(eps=eps, min_samples=min_samples)
    # predict upon DBSCAN and input points
    pred=ml_model.fit_predict(all_points)
    # draw 3D plot according to clustering reulst
    temp_ax.scatter(all_points[:, 0], all_points[:, 1], all_points[:,2], c=pred, s=20, cmap='jet')
    temp_ax.set_xlabel('f1')
    temp_ax.set_ylabel('f2')
    temp_ax.set_zlabel('f3')
    temp_ax.set_title("DBSCAN Clustering")
    for aa in range(0, 360):
            ax.view_init(angle, aa)
            plt.draw()
```

Now we run DBSCAN on augumented sets and visualize results.

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d') 
show_dbscan(zero_rotations,one_rotations,ax)
```

Compard with visualization of K-means, DBSCAN has better clustering performance. Although there still exists some misclassification, it can help seperate the overall distribution of "0" and "1" except some outliers.

Thus, the clustering performance show that the projection onto first 3 feature vectors after PCA can help extract acceptable features that separate the augumented image sets.

### 1.3.3 Feedforward Neural Network

In this section, we try neural network implementation. 

Since the task can be treated as simple binary classification problem, we use shallow network to implement our method.

After implementation, we find a feedforward net without back propogation is suffice to extract 2 features that help to separate 2 image sets.

(no need to use convolution layer in this task)

We implement in Pytorch.

Pipeline:

1. Define 1-layer neural network model, including fc layer

2. run the 1-layer nn upon augumented image sets and calculate the features

3. visualize the features upon a 1-layer nn

4. impove the model by 3-layer neural network model, including fc layer, activatoin function and softmax prediction

5. run the improved model nn upon augumented image sets and calculate the features

6. visualize the features upon improved model

7. visualize the prediction performance of improved model

```python
a=1
```

```python
import torch
import torch.nn.functional as F
from torch.utils.data import DataLoader
```

```python
# based on neural network model
# run the model by image sets
# calculate 2 features
# return np array of features
def get_nn_feature(images):
    # define neural network model
    model=Net()
    features=[]
    # transform into dataloader
    data_loader = DataLoader(dataset=images)
    for temp_image in data_loader:
        # resize each image for later input
        temp_image.resize_(784)
        temp_image=temp_image.float()
        # predict output
        outputs = model.forward(temp_image)
        outputs_array=outputs.data.numpy()
        # store outputs to array
        temp_feature_1=outputs_array[0]
        temp_feature_2=outputs_array[1]
        #temp_feature_3=outputs_array[2]
        features=features+[[temp_feature_1,temp_feature_2]]
    features=np.array(features)
    return features
```

```python
# visualize the 2 feature after running nn model
def draw_nn_features(images_1,images_2,temp_ax):
    # get feature values after running nn model
    zero_features=get_nn_feature(images_1)
    ones_features=get_nn_feature(images_2)
    # visualize the 2 features by scatter plot
    temp_ax.scatter(zero_features[:, 0], zero_features[:, 1], c='red',label='zeros')
    temp_ax.scatter(ones_features[:, 0], ones_features[:, 1], c='green',label='ones')
    temp_ax.set_xlabel('feature 1')
    temp_ax.set_ylabel('feature 2')
    temp_ax.set_title("Features after Fully Connected Layer")
    temp_ax.legend()
```

```python
# define a 1-layer fully conncected network
# the model will output tensor of 2 feature values
class Net(torch.nn.Module):
    def __init__(self ):
        super(Net, self).__init__()
        self.fc1 = torch.nn.Linear(784, 2)        
    def forward(self, x):
        #x = x.view(1,-1)
        x = self.fc1(x)
        return x
    
    def predict(self, x):
        logits = self.forward(x)
        return F.softmax(logits)
```

Now, we plot the 2D scatter plot of 2 features after running 1-layer fc model.

To augument original input image sets, we used the image sets after rotations. This can also help prove the invariance to rotation. The analysis of scaling and translation is same as it is in 1.2, thus we do not give further details on this 2 operations downwards.

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
draw_nn_features(zero_rotations,one_rotations,ax)
ax.set_title("Classification Predictions after 1 Fully Connected Layer")
```

The result depends on the initialization of weitht parameters of the net. Here We choose a representative classification result after several runs here.

It can be observed that there is an obvious distribution area of 2 image sets after running a 1-layer fc model, although there is overlapping in the center of plane, where the seperating is ambiguous.

```python
# define a 3-layer fully conncected network
# with activation functions of relu
# the model will output tensor of 2 feature values
# predict by softmax
class Net(torch.nn.Module):
    def __init__(self ):
        super(Net, self).__init__()
        self.fc1 = torch.nn.Linear(784, 200)
        self.fc2 = torch.nn.Linear(200, 20)
        self.fc3 = torch.nn.Linear(20, 2)
        
    def forward(self, x):
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        x = F.relu(x)
        x = self.fc3(x)
        return x
    
    def predict(self, x):
        logits = self.forward(x)
        return F.softmax(logits)
```

Now, we plot the 2D scatter plot of 2 features after running improved model of 3-layer network.

To augument original input image sets, we used the image sets after rotations. This can also help prove the invariance to rotation. The analysis of scaling and translation is same as it is in 1.2, thus we do not give further details on this 2 operations downwards

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
draw_nn_features(zero_rotations,one_rotations,ax)
ax.set_title("Classification Predictions after 3 Fully Connected Layer")
```

Although the result depends on the initialization of weitht parameters of the net, the overall performance of 3-layer network is obviously improved.

Here we choose a representative classification result that shows the good performance after running a 3-layer network.

Now we turn to the classification performance of network prediction.

```python
# based on improved neural network model
# run the model by image sets
# calculate softmax prediction of each image
# return np array of predictions
def get_nn_predictions(images):
    predictions=[]
    # improved neural network model
    model=Net()
    # transform into dataloader
    data_loader = DataLoader(dataset=images)
    for temp_image in data_loader:
        # resize each image for later input
        temp_image.resize_(784)
        temp_image=temp_image.float()
        outputs = model.forward(temp_image)
        predicted=model.predict(temp_image)
        # store predictions to array
        predicted_array=predicted.data.numpy()
        temp_pred_1=predicted_array[0]
        temp_pred_2=predicted_array[1]
        predictions=predictions+[[temp_pred_1,temp_pred_2]]
    predictions=np.array(predictions)
    return predictions
```

```python
# # visualize the 2 class predictions after running nn model
def draw_nn_predictions(images_1,images_2,temp_ax):
    # get predictions values after running nn model
    zero_predictions=get_nn_predictions(images_1)
    ones_predictions=get_nn_predictions(images_2)
    # visualize the 2 class predictions by scatter plot
    temp_ax.scatter(zero_predictions[:, 0], zero_predictions[:, 1], c='red',label='zeros')
    temp_ax.scatter(ones_predictions[:, 0], ones_predictions[:, 1], c='green',label='ones')
    temp_ax.set_xlabel('Class 1')
    temp_ax.set_ylabel('Class 2')
    temp_ax.legend()
```

Now, we plot the 2D scatter plot of 2 predictions after running improved model of 3-layer network.


```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))

draw_nn_predictions(zero_rotations,one_rotations,ax)
```

Although the result depends on the initialization of weitht parameters of the net, the overall predcition of a feedforward 3-layer network is good. 

Here we we choose a good classification result after several runs. In most runs, there are very few misclassifications.

Thus, in our task, we can say a feedforward neural network with 3 fully connected layer without backpropogation has good performance in extracting image features and predicting the class.
